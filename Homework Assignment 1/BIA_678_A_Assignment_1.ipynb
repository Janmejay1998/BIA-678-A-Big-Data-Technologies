```
%sql
DROP TABLE IF EXISTS diamonds;
 
CREATE TABLE diamonds USING CSV OPTIONS (path "/databricks-datasets/Rdatasets/data-001/csv/ggplot2/diamonds.csv", header "true")
```

```
diamonds = (spark.read
  .format("csv")
  .option("header", "true")
  .option("inferSchema", "true")
  .load("/databricks-datasets/Rdatasets/data-001/csv/ggplot2/diamonds.csv")
)
```

```
%sql
SELECT COUNT(*) FROM DIAMONDS
```

```
%sql
SELECT * FROM diamonds LIMIT 10
```

```
%sql
SELECT color, avg(price) AS Price_Col FROM diamonds GROUP BY color ORDER BY COLOR DESC
```

```
# diamonds.head(5)
diamonds.printSchema()
```

```
from pyspark.sql.functions import avg, col, desc
diamonds.groupBy("color").agg(avg("price").alias("Price_Col")).sort(desc("Price_Col")).show()
```
